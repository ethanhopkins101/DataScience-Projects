## Encode and Preparation

```python
def encode_categorical_columns(df, ordered_variables, nominal_columns):
    
    # Combine the nominal variables with the keys of the ordered_variables dictionary
    categorical_columns = nominal_columns + list(ordered_variables.keys())

    # Fill NaN values in ordinal and nominal columns with "Missing_Value"
    for col in categorical_columns:
        df[col].fillna("Missing_Value", inplace=True)

    # Add "Missing_Value" level for missing values in each ordered variable
    for col, levels in ordered_variables.items():
        levels_with_missing = ["Missing_Value"] + levels
        df[col] = pd.Categorical(df[col], categories=levels_with_missing, ordered=True)
    
    # Convert nominal categorical variables to nominal data type
    for col in nominal_columns:
        df[col] = pd.Categorical(df[col], ordered=False)
    
    return df, categorical_columns
```

**As it exhibits a non-normal probability distribution, it's crucial to conduct variable transformation tests to assess the attainment of a more 'normalized' distribution** 📊

We will test three different transformations (square root, logarithm, and Box-Cox) on the 'SalePrice' variable, and then it will display a boxplot alongside a histogram for each of them. Additionally, it will perform a normality test (e.g., Shapiro-Wilk test) after each transformation and show the p-value. The transformation that results in a p-value closer to 1 will indicate a more normal distribution. 📊🔄📈


```python
# Assuming 'df' is your DataFrame with the 'SalePrice' variable

# Define a list of transformation names and corresponding functions
transformations = {
    'Original': lambda x: x,
    'Square Root': np.sqrt,
    'Logarithm': np.log1p
}

# Prepare a color palette for plotting
colors = sns.color_palette('Set2', n_colors=len(transformations))

# Create subplots for boxplots and histograms
fig, axes = plt.subplots(nrows=len(transformations), ncols=2, figsize=(12, 8))

for i, (transformation_name, transformation_func) in enumerate(transformations.items()):
    # Apply the transformation to 'SalePrice'
    transformed_price = transformation_func(df_train['SalePrice'])
    
    # Plot the boxplot
    sns.boxplot(x=transformed_price, ax=axes[i, 0], color=colors[i])
    axes[i, 0].set_title(f'Boxplot - {transformation_name}')
    
    # Plot the histogram
    sns.histplot(transformed_price, ax=axes[i, 1], kde=True, color=colors[i])
    axes[i, 1].set_title(f'Histogram - {transformation_name}')
    
    # Perform a normality test (Shapiro-Wilk) and display the p-value
    p_value = stats.shapiro(transformed_price)[1]
    axes[i, 1].text(0.6, 0.8, f'p-value: {p_value:.4f}', transform=axes[i, 1].transAxes)

plt.tight_layout()
plt.show()
```

**As you can observe, the most effective transformation is the logarithmic one. It successfully normalizes the probability distribution of SalePrice in the best possible manner**

```python
SalePrice = np.log1p(df_train["SalePrice"])
df_train["SalePrice"] = SalePrice

not_nan_indices = df_all['SalePrice'].notna()
df_all.loc[not_nan_indices, 'SalePrice'] = SalePrice
```

## Numerical data transform

```python
#Numerical cols
numerical_columns = [col for col in df_all.columns.to_list() if col not in categorical_columns]
numerical_columns.remove('Id')
numerical_columns.remove('SalePrice')
```

```python
def find_best_transformation(df, column_name):
    """
    Find the best transformation for a numerical variable based on normality tests.

    Parameters:
    - df: DataFrame
    - column_name: str, the name of the numerical column to be transformed

    Returns:
    - transformed_column: Series, the transformed column with the best method
    """

    # Dictionary to store p-values for each transformation
    p_values = {}

    # Apply Box-Cox transformation
    try:
        transformed_column, lambda_value = boxcox(df[column_name] + 1e-10)
        p_value = stats.shapiro(transformed_column)[1]
        p_values['Box-Cox'] = p_value
    except Exception:
        pass

    # Apply log transformation
    try:
        transformed_column = np.log(df[column_name] + 1e-10)
        p_value = stats.shapiro(transformed_column)[1]
        p_values['Logarithm'] = p_value
    except Exception:
        pass

    # Apply square root transformation
    transformed_column = np.sqrt(df[column_name] + 1e-10)
    p_value = stats.shapiro(transformed_column)[1]
    p_values['Square Root'] = p_value

    # Find the best transformation based on the lowest p-value
    best_transformation = min(p_values, key=p_values.get)

    # Print the results
    print(f"Best transformation for {column_name}: {best_transformation} (p-value: {p_values[best_transformation]:.4f})")

    # Plot the original and transformed data
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 4))

    # Plot transformed data in red
    sns.histplot(transformed_column, kde=True, color='red', ax=axes[0])
    axes[0].set_title(f'Transformed Data - {best_transformation}')

    # Plot box-plot for transformed data in soft green
    sns.boxplot(x=transformed_column, color='purple', ax=axes[1])
    axes[1].set_title('Box-Plot - Transformed Data')    

    plt.tight_layout()
    plt.show()

    return transformed_column
```

```python
# Create a DataFrame to store normality test results
normality_results = pd.DataFrame(columns=['Variable', 'P-Value', 'Is Normal'])

# Loop to generate the plots and perform normality tests
for col in numerical_columns:

    # Perform a normality test (Shapiro-Wilk)
    _, p_value = stats.shapiro(df_all[col])
    alpha = 0.05  # Significance level
    is_normal = bool(p_value > alpha)  # Explicitly cast to bool dtype

    # Add the results to the DataFrame
    normality_results.loc[len(normality_results)] = [col, p_value, is_normal]

# Print the normality test results DataFrame
print(normality_results)
```

```python
warnings.filterwarnings("ignore", category=RuntimeWarning, module="numpy")
warnings.filterwarnings("ignore", category=RuntimeWarning, module="scipy.optimize")

#Find the best transformation to numerical cols
subset = normality_results[normality_results['Is Normal'] == False]
subset
for col in subset['Variable']:
    find_best_transformation(df_all, col)
```

### Qualitative Analysis

#### Definition of ANOVA (Analysis of Variance)

An F-test is any statistical test in which the test statistic has an F-distribution under the null hypothesis. It is most often used when comparing statistical models that have been fitted to a data set, in order to identify the model that best fits the population from which the data were sampled. Exact "F-tests" mainly arise when the models have been fitted to the data using least squares. The name was coined by George W. Snedecor, in honour of Ronald Fisher. Fisher initially developed the statistic as the variance ratio in the 1920s. In summary, Analysis of Variance (ANOVA) is a statistical technique used to compare the means of three or more groups or categories. It is often used to test whether there are significant differences between the population means of distinct groups. It provides a structured approach to hypothesis testing and allows for deeper insights into group differences. However, it's crucial to use ANOVA appropriately, considering its underlying assumptions and conducting post-hoc tests when necessary, to draw valid conclusions from your data.

#### Mathematical Formulation


Let's assume you have $(k)$ groups with $(n_i)$ observations in each group, where $(i = 1, 2, ..., k)$.

The statistical hypotheses are formulated as:

- **Null Hypothesis $(H_0)$:** All population means are equal.
  
  $H_0: \mu_1 = \mu_2 = \ldots = \mu_k$

- **Alternative Hypothesis $(H_1)$:** At least one population mean is different from the others.
  
  $H_1: \text{At least one of the population means is different.}$

The F-test statistic is calculated as the ratio of between-group variability to within-group variability. The formula for the F-statistic is:

$F = \frac{\text{Between-Group Variability}}{\text{Within-Group Variability}}$

Where:

- **Between-Group Variability** is the sum of squared differences between the means of each group and the overall mean, weighted by the number of observations in each group.
  $\text{Between-Group Variability} = \sum_{i=1}^{k} n_i (\bar{X}_i - \bar{X})^2$
  
- **Within-Group Variability** is the sum of squared differences between each observation and the mean of the group to which it belongs.
  
  $\text{Within-Group Variability} = \sum_{i=1}^{k} \sum_{j=1}^{n_i} (X_{ij} - \bar{X}_i)^2$

Here, $(\bar{X}_i)$ represents the mean of group $(i)$, and $(\bar{X})$ is the overall mean.

Finally, the F-statistic follows an F-distribution with $(k-1)$ and $(N - k)$ degrees of freedom, where $(N)$ is the total number of observations.


**Reference:**
- [ANOVA - Wikipedia](https://en.wikipedia.org/wiki/Analysis_of_variance)





```python
warnings.filterwarnings("ignore", category=stats.DegenerateDataWarning)

# Define a function to perform ANOVA
def perform_anova(frame, target):
    # Create an empty DataFrame to store results
    anova_results = pd.DataFrame()
    anova_results['Feature'] = categorical_columns  # Assuming qualitative is predefined

    p_values = []

    # Iterate through qualitative features
    for feature in categorical_columns:
        # Create a list to store samples for each category
        samples = []

        # Iterate through unique categories within the feature
        for category in frame[feature].unique():
            # Get SalePrice values for the current category
            sale_prices = frame[frame[feature] == category][target].values
            samples.append(sale_prices)

        # Perform ANOVA test and get the p-value
        p_value = stats.f_oneway(*samples)[1]
        p_values.append(p_value)

    # Add p-values to the DataFrame
    anova_results['p-value'] = p_values

    # Sort the results by p-value
    anova_results = anova_results.sort_values(by='p-value')

    return anova_results

# Perform ANOVA on the 'train' DataFrame
result = perform_anova(df_train, 'SalePrice') #It's because in all the data there are missing values in SalePrice

# Calculate the disparity for visualization
result['Disparity'] = np.log(1.0 / result['p-value'].values + 1e-9)

# Create a barplot to visualize the results
plt.figure(figsize=(12, 6))
sns.barplot(data=result, x='Feature', y='Disparity')
plt.xticks(rotation=90)
plt.xlabel('Feature')
plt.ylabel('Disparity (log(1/p-value))')
plt.title('ANOVA Test Results')
plt.show()
```

### Quantitative Analysis

#### Correlation

Spearman, Kendall, and Pearson are correlation coefficients used to measure the relationship between variables. Spearman is ideal for assessing monotonic relationships in ordinal data or when a non-linear relationship is suspected, as it focuses on rank orders. Kendall's Tau is also suitable for ordinal data and assesses the concordance of rankings, making it robust to outliers and non-linear relationships. Pearson is best suited for linear relationships between continuous variables, assuming a normal distribution. It's sensitive to outliers and may not capture non-linear associations effectively. The choice depends on the nature of the data and the type of relationship under investigation, with Spearman and Kendall being preferable when data is not normally distributed or non-linearity is expected.

**Pearson Correlation Coefficient:**

*Mathematical Definition:* 

The Pearson correlation coefficient, denoted by $\rho$ (rho), measures the linear correlation between two continuous variables $XX$ and $YY$. It is given by the formula:

   $\rho = \frac{\sum{(X - \bar{X})(Y - \bar{Y})}}{\sqrt{\sum{(X - \bar{X})^2} \sum{(Y - \bar{Y})^2}}}$

Where:
- $\bar{X}$ is the mean of $X$.
- $\bar{Y}$ is the mean of $Y$..

These mathematical definitions elucidate the calculation process for each correlation coefficient, conveying the statistical relationship between variables $X$ and $Y$ in terms of rankings, linearity, or concordance, as applicable to each coefficient.

The chosen correlation is Pearson because it assumes a linear relationship between the independent variables and the SalePrice variable. 

```python
def calculate_and_plot_correlations(df, target_variable):
    # Calculate Pearson correlations between quantitative variables and the target variable
    correlations = {}
    for var in numerical_columns:
        if var != target_variable:
            correlations[var] = df[target_variable].corr(df[var], method='pearson')

    # Creating a DataFrame to store correlations
    correlation_df = pd.DataFrame({'Variable': list(correlations.keys()), 'Pearson Correlation': list(correlations.values())})

    # Sorting correlations in descending order
    correlation_df = correlation_df.sort_values(by='Pearson Correlation', ascending=False)

    # Creating a bar plot of correlations with improved grid and custom colors
    plt.figure(figsize=(10, 12))
    ax = sns.barplot(data=correlation_df, x='Pearson Correlation', y='Variable', palette='coolwarm_r')
    plt.title(f'Pearson Correlations with {target_variable}')
    plt.xlabel('Pearson Correlation')
    plt.ylabel('Variable')

    # Adding a grid
    ax.xaxis.grid(True, linestyle='--', alpha=0.6)
    ax.set_axisbelow(True)

    # Adjusting the size of variable names
    plt.xticks(fontsize=12)  # You can adjust the font size as needed

    plt.show()

    return correlation_df

# Call the function with your DataFrame and target variable
correlation_df = calculate_and_plot_correlations(df_all, 'SalePrice')
```

To fill in the missing values of GarageYrBlt based on the probability of proximity and the correlation with YearBuilt, you can use a linear interpolation approach. The idea is to create a linear function that relates the two variables and then use that function to fill in the missing values.

The general formula for a linear interpolation function is:

$y=mx+b$

Where:

- *y* is the value we want to interpolate (in this case, GarageYrBlt).
- *x* is the known value (YearBuilt).
- *m* is the slope of the line (correlation between the variables).
- *b* is the $y$-intercept (adjustment based on the probability of proximity).

```python
# Calculate the slope of the line (m)
slope = pearson_correlation

# Calculate the y-intercept (b)
intercept = probability_of_proximity * slope

df_all['GarageYrBlt'].fillna(df_all['YearBuilt'] * slope + intercept, inplace=True)

quanti_cols_to_fix.remove('GarageYrBlt')
```

The interpretation would be as follows:

- A high Full Bath Density indicates that there are many full bathrooms relative to the space, which can be seen as an indicator of convenience and comfort.
- A low Full Bath Density indicates that there are fewer full bathrooms relative to the space, which may indicate a home configuration with fewer bathrooms or less emphasis on easily accessible bathrooms.

```python
df_all['FullBathDensity'] = df_all['FullBath'] / df_all['GrLivArea']
```

An indicator that you can create based on the variables TotalBsmtSF (Total Basement Area) and 1stFlrSF (First Floor Area) is the "Basement Area to First Floor Area Ratio."

The formula for this indicator would be:

$Basement Area to First Floor Area Ratio =\frac{TotalBsmtSF}{1stFlrSF}$


This metric provides a measure of the relationship between the basement area and the first-floor area of a house. A higher ratio indicates that the basement is relatively larger compared to the first floor, while a lower ratio indicates the opposite.


Interpretation:

- If the basement area to first floor area ratio is close to 1, it suggests that the basement and the first floor have similar areas.
- A ratio less than 1 indicates that the first floor is larger than the basement in relation to area.
- A ratio greater than 1 indicates that the basement is larger in relation to the first floor.

```python
#Create BAFFAreaRatio
df_all['BAFFAreaRatio'] = df_all['TotalBsmtSF'] / df_all['1stFlrSF']

#Fill Missing
df_all['BAFFAreaRatio'].fillna(df_all['BAFFAreaRatio'].mean(), inplace=True)
```

### DecisionTreeClassifier to fill Categorical NaN values

- A chi-squared test (also chi-square or χ2 test) is a statistical hypothesis test used in the analysis of contingency tables when the sample sizes are large. In simpler terms, this test is primarily used to examine whether two categorical variables (two dimensions of the contingency table) are independent in influencing the test statistic (values within the table). 

- The test is valid when the test statistic is chi-squared distributed under the null hypothesis, specifically Pearson's chi-squared test and variants thereof. Pearson's chi-squared test is used to determine whether there is a statistically significant difference between the expected frequencies and the observed frequencies in one or more categories of a contingency table. For contingency tables with smaller sample sizes, a Fisher's exact test is used instead.

- In the standard applications of this test, the observations are classified into mutually exclusive classes. If the null hypothesis that there are no differences between the classes in the population is true, the test statistic computed from the observations follows a χ2 frequency distribution. The purpose of the test is to evaluate how likely the observed frequencies would be assuming the null hypothesis is true.

- Test statistics that follow a χ2 distribution occur when the observations are independent. There are also χ2 tests for testing the null hypothesis of independence of a pair of random variables based on observations of the pairs.

- Chi-squared tests often refers to tests for which the distribution of the test statistic approaches the χ2 distribution asymptotically, meaning that the sampling distribution (if the null hypothesis is true) of the test statistic approximates a chi-squared distribution more and more closely as sample sizes increase. 

```python
msz_na = df_all[df_all['MSZoning'] == 'Missing_Value'].copy(deep=True)
msz_notna = df_all[df_all['MSZoning'] != 'Missing_Value'].copy(deep=True)
```

```python
ct_dict = dict()
ct_result = dict()

for col in msz_notna:
    ct_dict[col] = pd.crosstab(msz_notna['MSZoning'], msz_notna[col])
    
for key, value in ct_dict.items():
    chi2, p, _, _ = chi2_contingency(value)
    if p < 0.05:
        ct_result[key] = [chi2, p]

chi2_result = pd.DataFrame.from_dict(ct_result, orient='index', columns=['chi2', 'p-value']).reset_index()
chi2_result.rename(columns={'index': 'Variable'}, inplace=True)
```

```python
chi2_result[chi2_result['p-value'] < 0.01]
```

- According to the chi-square test, all the variables are significant 
- A more refined criterion will be needed to verify the association, namely the Cramer-v criterion

The Cramér's V coefficient is a measure of association between categorical variables. It ranges from 0 to 1, where 0 means no association and 1 means total association. Here is the formula for the Cramér-V coefficient:

$V=\sqrt{x²/n*min(k-1, r-1)}$



where:

- $χ2$ is the Chi-Square value,
- $n$ is the total number of observations,
- $k$ is the number of columns in the contingency table,
- $r$ is the number of rows in the contingency table.

```python
cramer_v_list = []

for col in chi2_result['Variable']:
    n = np.sum(ct_dict[col].values)
    k = ct_dict[col].shape[1]
    r = ct_dict[col].shape[0]
    
    row = chi2_result[chi2_result['Variable'] == col]
    chi2_value = row['chi2'].values[0] if not row.empty else None
    
    if chi2_value is not None:
        cramer_v = np.sqrt(chi2_value / (n * min(k - 1, r - 1)))
        cramer_v_list.append(cramer_v)
    else:
        cramer_v_list.append(None)

chi2_result['cramer_v'] = cramer_v_list
```

```python
chi2_result_ready = chi2_result[chi2_result['cramer_v'] > 0.6]
```

```python
chi2_result_ready = chi2_result_ready[~chi2_result_ready['Variable'].isin(['MSZoning', 'SalePrice'])]
```

## Select Model

```python
from sklearn.svm import SVR
from sklearn.linear_model import ElasticNet
from sklearn.exceptions import DataConversionWarning
from sklearn.model_selection import cross_val_score, GridSearchCV
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
import heapq

warnings.filterwarnings("ignore", category=DataConversionWarning)
warnings.filterwarnings("ignore", category=RuntimeWarning)

# Define the models
models = {
    'Ridge': Ridge(),
    'Lasso': Lasso(),
    'ANN': MLPRegressor(max_iter=1000),  # Artificial Neural Networks (MLP)
    'XGBoost': XGBRegressor(),
    'GradientBoosting': GradientBoostingRegressor(),
    'RandomForest': RandomForestRegressor(),
    'SVR': SVR(),
    'ElasticNet': ElasticNet(),
}

# Use a pipeline with StandardScaler for scale-sensitive models
for model_name, model in models.items():
    models[model_name] = make_pipeline(StandardScaler(), model)

# Perform cross-validation for each model
model_rmse = {}

for model_name, model in models.items():
    scores = cross_val_score(model, X_train, y_train, cv=5, scoring='neg_mean_squared_error')
    rmse_scores = np.sqrt(-scores)
    
    # Calculate the mean RMSE from cross-validation
    mean_rmse = np.mean(rmse_scores)
    model_rmse[model_name] = mean_rmse
    
    print(f"Model: {model_name}")
    print(f"Mean RMSE from cross-validation: {mean_rmse:.4f}\n")

# Choose the models with the three lowest mean RMSE in cross-validation using GridSearchCV
param_grids = {
    'Ridge': {'ridge__alpha': [0.1, 0.93, 10]},
    'Lasso': {'lasso__alpha': [0.1, 0.00244, 10]},
    'ANN': {'mlpregressor__alpha': [0.0001, 0.001, 0.01], 'mlpregressor__hidden_layer_sizes': [(100,), (50, 50)]},
    'XGBoost': {'xgbregressor__learning_rate': [0.04, 0.1, 0.2], 'xgbregressor__max_depth': [3, 4, 5]},
    'GradientBoosting': {'gradientboostingregressor__learning_rate': [0.01, 0.1, 0.2], 'gradientboostingregressor__max_depth': [3, 4, 5]},
    'RandomForest': {'randomforestregressor__n_estimators': [50, 200, 570]},
    'SVR': {'svr__C': [0.1, 1, 10], 'svr__epsilon': [0.1, 0.3, 0.5]},
    'ElasticNet': {'elasticnet__alpha': [0.1, 1, 10], 'elasticnet__l1_ratio': [0.3, 0.5, 0.7]},
    # ...
}

top_models = []

for model_name, model in models.items():
    grid_search = GridSearchCV(model, param_grids[model_name], cv=5, scoring='neg_mean_squared_error')
    grid_search.fit(X_train, y_train)
    
    # Choose the best model after grid search
    best_model = grid_search.best_estimator_
    
    # Calculate the mean RMSE from cross-validation with best hyperparameters
    mean_rmse = np.sqrt(-grid_search.best_score_)
    model_rmse[model_name] = mean_rmse
    
    print(f"Model: {model_name}")
    print(f"Best hyperparameters: {grid_search.best_params_}")
    print(f"Mean RMSE from cross-validation with best hyperparameters: {mean_rmse:.4f}\n")
    
    # Keep track of the three best models
    if len(top_models) < 3:
        heapq.heappush(top_models, (mean_rmse, model_name, best_model))
    else:
        heapq.heappushpop(top_models, (mean_rmse, model_name, best_model))

# Train the selected models on the full training set
for _, model_name, best_model in top_models:
    best_model.fit(X_train, y_train)

# Predict values on the test set for the top models
predictions = []

for _, _, best_model in top_models:
    y_pred = best_model.predict(X_test)
    predictions.append(y_pred)

# Calculate the RMSE on the test set for the top models
for i, (_, model_name, _) in enumerate(top_models):
    test_rmse = np.sqrt(mean_squared_error(y_test, predictions[i]))
    print(f"Model {i + 1} ({model_name}): RMSE on the test set: {test_rmse:.4f}\n")

# Plot predictions of the top models
for i, (_, model_name, _) in enumerate(top_models):
    plt.figure(figsize=(8, 6))
    plt.title(f'Predictions - Model {i + 1} ({model_name})', fontsize=16)
    plt.xlabel('Example', fontsize=14)
    plt.ylabel('Real Value vs. Predicted Value', fontsize=14)

    plt.plot(y_test, label='Real', linestyle='--')
    plt.plot(predictions[i], label='Predicted')

    plt.legend()
    plt.show()
```
